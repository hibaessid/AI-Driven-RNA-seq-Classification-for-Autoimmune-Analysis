# Data Overview and Preprocessing

# common genes (features) between all datasets

```python
# Get the common genes (features) between all datasets
common_genes = ra_train.columns.intersection(sle_train.columns).intersection(healthy_train.columns)

# Reduce each dataset to only the common genes
ra_train = ra_train[common_genes]
ra_test = ra_test[common_genes]

sle_train = sle_train[common_genes]
sle_test = sle_test[common_genes]

healthy_train = healthy_train[common_genes]
healthy_test = healthy_test[common_genes]

# check the new shapes of the datasets
print("RA Train Shape:", ra_train.shape)
print("SLE Train Shape:", sle_train.shape)
print("Healthy Train Shape:", healthy_train.shape)
```

# Handling Class Imbalance Using SMOTE (Synthetic Minority Over-sampling Technique

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separate features and labels
X_train = combined_train_shuffled.drop(['label', 'Unnamed: 0'], axis=1) # Drop 'Unnamed: 0' column
y_train = combined_train_shuffled['label']

# Check original class distribution
print(f"Original class distribution: {Counter(y_train)}")

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check new class distribution after applying SMOTE
print(f"New class distribution: {Counter(y_train_smote)}")


# Standardize the resampled data
scaler = StandardScaler()
X_train_smote_scaled = scaler.fit_transform(X_train_smote)
```

# log transformation for reducing skewness in the RNA-seq data

```python
# Check for negative values in the training data
negative_values_train = (X_train_smote_scaled_cleaned_df < 0).any().sum()
print(f"Number of features with negative values in the training data: {negative_values_train}")

# Check for negative values in the test data
negative_values_test = (X_test_scaled_cleaned_df < 0).any().sum()
print(f"Number of features with negative values in the test data: {negative_values_test}")
```

```python
# Find the minimum value in the training and test datasets
min_value_train = X_train_smote_scaled_cleaned_df.min().min()
min_value_test = X_test_scaled_cleaned_df.min().min()

# Calculate the shift needed to make all values positive
shift_value = abs(min(min_value_train, min_value_test)) + 1e-6  # Add a small epsilon to avoid exact zero

# Apply the shift to the datasets
X_train_shifted = X_train_smote_scaled_cleaned_df + shift_value
X_test_shifted = X_test_scaled_cleaned_df + shift_value
```

```python
# Apply log1p transformation to the shifted data
X_train_log_transformed = np.log1p(X_train_shifted)
X_test_log_transformed = np.log1p(X_test_shifted)

# Convert back to DataFrame for easier manipulation and visualization later
X_train_log_transformed = pd.DataFrame(X_train_log_transformed, columns=X_train_shifted.columns)
X_test_log_transformed = pd.DataFrame(X_test_log_transformed, columns=X_test_shifted.columns)

X_train_log_transformed.head()
```

# perform dimensionality reduction using PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA

# Instantiate the PCA model
pca = PCA(n_components=0.95)  # Retain 95% of the variance

# Fit PCA on the training data and transform it
X_train_pca = pca.fit_transform(X_train_log_transformed)

# Apply the same transformation to the test data
X_test_pca = pca.transform(X_test_log_transformed)

# Check the shape of the reduced dataset
print(f"Shape of PCA-transformed training data: {X_train_pca.shape}")
print(f"Shape of PCA-transformed test data: {X_test_pca.shape}")
```

# Feature Selection Using SelectKBest (ANOVA F-value)

```python
# Import necessary libraries
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

# Number of top features to select (adjust based on model performance)
k = 100

# Convert y_train_smote to a NumPy array if it is a DataFrame
y_train_smote_array = y_train_smote.to_numpy().ravel()

# Instantiate SelectKBest with f_classif as the score function
selector = SelectKBest(f_classif, k=k)  # Create an instance of SelectKBest and assign it to selector

# Fit the selector to the PCA-transformed training data with reshaped y_train_smote
X_train_selected = selector.fit_transform(X_train_pca, y_train_smote_array)

# Apply the same transformation to the test data
X_test_selected = selector.transform(X_test_pca)

# Get a list of the column names before PCA
original_features = X_train_log_transformed.columns.tolist()

# Get the indices of the selected features after PCA
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features from the original feature list
selected_features = [original_features[i] for i in selected_indices]

# Convert the selected features back to DataFrames with gene names
X_train_selected = pd.DataFrame(X_train_selected, columns=selected_features)
X_test_selected = pd.DataFrame(X_test_selected, columns=selected_features)

# Check the shape of the reduced dataset
print(f"Shape of the training data after SelectKBest: {X_train_selected.shape}")
print(f"Shape of the test data after SelectKBest: {X_test_selected.shape}")
```

# Modeling Phase

# Split the data into train and validation

```python
from sklearn.utils import shuffle

# Shuffle the data
X_train_selected_shuffled, y_train_smote_array_shuffled = shuffle(X_train_selected, y_train_smote_array, random_state=42)

# Split the shuffled data into train (80%) and validation (20%) sets
from sklearn.model_selection import train_test_split

X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train_selected_shuffled, y_train_smote_array_shuffled, test_size=0.2, random_state=42
)

# Display the shape of the split data
print(f"Training set shape: {X_train_final.shape}, Training labels shape: {y_train_final.shape}")
print(f"Validation set shape: {X_val.shape}, Validation labels shape: {y_val.shape}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import make_scorer, f1_score

# Instantiate the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

# Define the parameter grid
param_distributions = {
    'n_estimators': [50, 100, 200, 300, 400],
    'max_depth': [3, 4, 5, 6],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9],
    'gamma': [0, 0.1, 0.2, 0.3],
    'reg_alpha': [0, 0.1, 0.5, 1],
    'reg_lambda': [0.5, 1, 1.5, 2]
}

# Set up the RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_distributions,
    n_iter=50,  # Number of parameter combinations to try
    scoring=make_scorer(f1_score, average='weighted'),  # Use weighted F1-score due to class imbalance
    cv=5,  # 5-fold cross-validation
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit the randomized search on the training data
random_search.fit(X_train_final, y_train_final)

# Print the best parameters and best score found
print("Best parameters found: ", random_search.best_params_)
print("Best cross-validation f1-score: {:.2f}".format(random_search.best_score_))
```
